```python
def gsheet_to_tabular(input_str):
    latex_table = ""

    lines = input_str.strip().split('\n')
    num_cols = len(lines[0].split('\t'))

    data = []
    for line in lines:
        columns = line.strip().split('\t')
        data.append([val if idx == 0 else round(float(val.rstrip('%')), 3) for idx, val in enumerate(columns)])

    min_values = [min(row[col_idx] for row in data) for col_idx in range(1, num_cols)]

    for row_idx, row in enumerate(data):
        latex_table += "& "
        for col_idx, val in enumerate(row):
            if col_idx == 0:
                latex_table += str(val)
            else:
                if val == min_values[col_idx - 1]:
                    latex_table += "\\textbf{" + str(val) + "\%}"  # Add '%' back
                else:
                    latex_table += str(val) + "\%"  # Add '%' back
            if col_idx < num_cols - 1:
                latex_table += " & "
        latex_table += " \\\ \n"

    return latex_table
```
